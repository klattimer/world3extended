# World3 Extended Exploratory Analysis
This notebook runs baseline and polycrisis scenarios and compares systemic stability trajectories.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd

from world3.model import SimulationConfig, World3Model
from world3.utils.metrics import detect_collapse
```

```python
cfg = World3Model.load_yaml(Path('../config/default.yaml'))
sim = SimulationConfig.from_mapping(cfg)

baseline = World3Model(sim, cfg, scenario_name='baseline').run()
poly = World3Model(sim, cfg, scenario_name='polycrisis').run()

baseline[['year', 'systemic_stability', 'energy_shortage', 'food_stress']].head()
```

```python
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(baseline['year'], baseline['systemic_stability'], label='baseline', lw=2)
ax.plot(poly['year'], poly['systemic_stability'], label='polycrisis', lw=2)
ax.set_title('Systemic Stability Comparison')
ax.set_ylabel('Stability Index')
ax.grid(alpha=0.3)
ax.legend()
plt.show()
```

```python
collapse_baseline = detect_collapse(baseline)
collapse_poly = detect_collapse(poly)

collapse_summary = pd.DataFrame({
    'scenario': ['baseline', 'polycrisis'],
    'collapsed': [bool(collapse_baseline.any()), bool(collapse_poly.any())],
    'collapse_year': [
        int(baseline.loc[collapse_baseline, 'year'].min()) if collapse_baseline.any() else None,
        int(poly.loc[collapse_poly, 'year'].min()) if collapse_poly.any() else None,
    ],
})
collapse_summary
```
